```python
import pandas as pd

df = load_menu2160_snapshot(date_ref='2025-02-13')

def get_df_preprocessed_column_menu2160(menu2160):
    col1 = menu2160.columns
    col2 = list(menu2160.iloc[0])
    data = {'col1':col1, 'col2':col2}
    df = pd.DataFrame(data)
    df['col1'] = df['col1'].apply(lambda x: x.replace('\n','').split('.')[0])
    df['col1'] = df['col1'].apply(lambda x: None if 'Unnamed:' in x else x)
    df['col2'] = df.apply(lambda x: x['col1'] if x['col1']=='증감' else x['col2'], axis=1)
    df['col2'] = df['col2'].apply(lambda x: '' if pd.isna(x) else x)
    df['col1'] = df['col1'].apply(lambda x: None if x=='증감' in x else x).ffill()
    df['col'] = df.apply(lambda x: f"{x['col1']}: {x['col2']}" if x['col2']!='' else x['col1'], axis=1)
    return df

def preprocess_column_menu2160(menu2160):
    df = get_df_preprocessed_column_menu2160(menu2160)
    col = list(df['col'])
    return col
```

```python
df = load_menu2160_snapshot(date_ref='2025-02-13')
preprocess_column_menu2160(df)
```
